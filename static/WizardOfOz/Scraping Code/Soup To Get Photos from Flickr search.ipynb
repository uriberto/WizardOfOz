```python
import bs4 as bs
import urllib.request
```

```python
with open('The Road To Oz Links') as f:
    lines = f.read()
```

```python
f.close
lines
```

```python
soup = bs.BeautifulSoup(lines,'lxml')
```

```python
link_list = soup.find_all("a", {"class": "overlay"})
```

```python
href_list = ['http://flickr.com'+linkitem['href']+'sizes/o' for linkitem in link_list]

print(href_list)
```

```python
img_src_list = []
for url in href_list:    
    asoup = bs.BeautifulSoup(urllib.request.urlopen(url).read(),'lxml')
    div = asoup.find('div', {"id":'allsizes-photo'})
    img = div.find('img')
    img_src_list.append(img['src'])
print(img_src_list)
```

```python
import requests # to get image from the web
import shutil # to save it locally

def download_img(image_url):
    ## Set up the image URL and filename
    filename = image_url.split("/")[-1]

    # Open the url image, set stream to True, this will return the stream content.
    r = requests.get(image_url, stream = True)

    # Check if the image was retrieved successfully
    if r.status_code == 200:
        # Set decode_content value to True, otherwise the downloaded image file's size will be zero.
        r.raw.decode_content = True

        # Open a local file with wb ( write binary ) permission.
        with open(filename,'wb') as f:
            shutil.copyfileobj(r.raw, f)

        print('Image sucessfully Downloaded: ',filename)
    else:
        print('Image Couldn\'t be retreived')
```

```python
for url in img_src_list:
    download_img(url)
```
